```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
# Definimos el polinomio
def polinomio(x):
    return x**5 - 6*x**4 + 2*x**3 + 20*x**2 - 27*x + 10

valores_iniciales = [-10, -5, 0, 5, 10]

# Encontramo las raíces usando fsolve
raices = fsolve(polinomio, valores_iniciales)

raices_unicas = np.unique(np.round(raices, decimals=5))

print("Raíces encontradas :")
for i, raiz in enumerate(sorted(raices_unicas), start=1):
    print(f"x{i} = {raiz}")
```

```python
# Creación de la gráfica
x = np.linspace(-12.5, 12.5, 400)
y = polinomio(x)
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Polinomio $x^5 - 6x^4 + 2x^3 + 20x^2 - 27x + 10$', color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.scatter(raices_unicas, polinomio(raices_unicas), color='red', zorder=5, label='Raíces')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfica del Polinomio y sus Raíces')
plt.legend()
plt.grid(True)
plt.show()
```
